```python
import matplotlib.pylab as plt
import cv2
import numpy as np
```

```python
image = cv2.imread('/Users/saishashetty/Downloads/image.jpg')

#converting the image into RGB format because we are going to load this image using matplotlib
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.show()

# observing the image below we see that that horizontal range is from 0 to 1200. However when we observe the vertical line we see that the range increases from top to bottom unlike the other graphs.This happens in matplotlib presentation
```

```python
print(image.shape)
height = image.shape[0]
width = image.shape[1]

#The ROIV is the three points we will plot.

region_of_interest_vertices = [
    (0, height),
    (width/2, height/2),
    (width, height)
]
```

```python
def region_of_interest(img, vertices):
    mask = np.zeros_like(img)  #This takes one parameter that is img
    channel_count = img.shape[2] #we saw in image shape we got 3 values i.e width and height
    match_mask_color = (255,) * channel_count  #Creating a match color with the same colour of channel_count
    cv2.fillPoly(mask, vertices, match_mask_color)  #We need to fill the polygon
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

cropped_image = region_of_interest(image,
                np.array([region_of_interest_vertices], np.int32),)
plt.imshow(cropped_image)
plt.show()
```
